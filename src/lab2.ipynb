```python
# Word2Vec
import gensim.downloader as api
import time
```

```python
# загружаем предобученную модель
wv = api.load('word2vec-ruscorpora-300')
```

```python
# создание списка слов без частей речи
vocab = []
for w in list(wv.index_to_key):
  vocab.append(w.split('_')[0])
```

```python
# Список векторов текстов для всего корпуса
vectors_corpus = []

# Для отслеживания прогресса выполнения
count = 0

# Время начала выполнения цикла. Для подсчета времени выполнения
start = time.time()

for x in range (1260):

    # Каждые 100 текстов выводится информация об количестве обработанных текстов
    if count % 100 == 0:
        print(count)

    # Выбор текущей строки из датафрейма
    preprocessing_text_list = df.loc[x]['preprocessed_text'].split()

    # Список векторов текущего текста
    vectors_word = []

    for i in preprocessing_text_list:
        if i in vocab:
            vectors_word.append(wv[list(wv.index_to_key)[vocab.index(i)]])
        else:
            vectors_word.append(np.zeros(300))

    # Для отследивания прогресса
    count += 1

    # Высчитывание среднего вектора для текущего текста по векторам каждого слова
    vector_text = np.mean(vectors_word, axis = 0)

    # Добавление вектора текущего текста в список векторов корпуса
    vectors_corpus.append(vector_text)

# Время окончания выполнения цикла
end = time.time()

# Вывод времени выполнения цикла
print('It took:', end - start, 'seconds')
```

```python
df_w2v = pd.DataFrame(vectors_corpus)
```

```python
df_w2v.shape
```

```python
df_w2v.to_csv('df_w2v.csv', index=False)
```

### 3. Разделите отзывы на положительные и отрицательные можно на основании значения в поле «score» (например, 1,2- негативные, 4,5-позитивные).

```python
# Разделение на положительные и отрицательные отзывы

def assign_labels(scores):
    threshold = 3  # пороговое значение
    labels = []

    for score in scores:
        if score >= threshold:
            labels.append(1)  # Отзыв является положительным
        else:
            labels.append(0)  # Отзыв является отрицательным

    return labels
```

### 4. Классифицируйте отзывы на положительные/отрицательные с использованием разных векторных представлений. Можно использовать любую модель/модели класссификации.

**Логистическая регрессия**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

*   BoW



```python
X = df_BOW.values
Y = df['score']
```

```python
y_labels = assign_labels(Y)
```

```python
# разделением наш корпус данных на обучающую и тестовую выборку
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_labels,
    test_size=0.2,
    shuffle = True,
    random_state=42,
    stratify = None)
```

```python
# классификатор
clf = LogisticRegression(
        penalty = 'l2', #{‘l1’, ‘l2’, ‘elasticnet’, None}, default=’l2’
        C = 0.2, #float, default=1.0
        class_weight = 'balanced', #dict or ‘balanced’, default=None
        random_state = 42,
        max_iter = 100, #int, default=100
        n_jobs = -1, #int, default=None
)
# предобученный классификатор
clf_pretrained = clf.fit(X_train, y_train)
```

```python
# результат предсказания
y_predict = clf_pretrained.predict(X_test)
y_predict
```

```python
# Оценка точности
print(classification_report(y_test, y_predict, digits = 4))
```
